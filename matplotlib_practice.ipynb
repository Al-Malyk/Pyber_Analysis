```python
# import dependencies
import matplotlib.pyplot as plt
```

```python
# x axis set to a list of strings
x_axis = ['Jan' ,'Feb' , 'Mar' , 'April' , 'May', 'June' , 'July' , 'Aug' , 'Sept' , 'Oct' , 'Nov' , 'Dec']

# y_axis set to the prices of fares
y_axis = [10.02 , 23.24, 39.20 , 35.42, 32.34 , 27.04 , 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.plot(x_axis , y_axis)
plt.xlabel('Months')
plt.ylabel('Fares')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis , y_axis, label = 'Boston',color = "green" , linewidth = 2)
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_ylim(0,45)
ax.grid()
```

```python
# create the bar chart 
plt.bar(x_axis , y_axis, color = "blue" , label="Boston")
plt.xlabel('Months')
plt.ylabel('Fares')
plt.title('PyBer Fare by Month')
plt.legend()
plt.show()
```

```python
# create an horizontal bar chart 
plt.barh(x_axis , y_axis, color = "purple" , label="Boston")
plt.gca().invert_yaxis()
plt.xlabel('Fares ($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.legend()
plt.show()
```

```python
fig ,ax = plt.subplots()
ax.barh(x_axis , y_axis, color='cyan', label='Chicago')
ax.invert_yaxis()
ax.legend()
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
```

```python
# create a scatterplot
#  plt.plot(x_axis , y_axis, 'o') type in the 'o' to make this a scatter plot and not a line graph
plt.scatter(y_axis , x_axis, s = [i*3 for i in y_axis] ,color="red" , label = 'Chicago')
plt.gca().invert_yaxis()
plt.xlim(0,45)
plt.title('PyBer Fare by Month')
plt.xlabel('Date')
plt.ylabel('Fares($)')
plt.legend()
```

```python
fig , ax = plt.subplots()
ax.scatter(y_axis, x_axis,s = [i*5 for i in y_axis], color='skyblue', edgecolor='black' ,label='Boston',linewidth=2,alpha=0.8)
ax.invert_yaxis()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel('Date')
ax.set_xlim(0,45)
ax.set_ylabel('Fare($)')
ax.legend()
```

```python
# create a pie chart
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0,)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, labels=x_axis, autopct='%.1f%%',shadow=True ,explode=explode_values)

plt.show()
```

```python
fig , ax = plt.subplots(figsize=(8,8))
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
ax.pie(y_axis ,labels=x_axis, autopct="%.1f%%", explode=explode_values, shadow=True, startangle=95, counterclock=False)
# did a bit of trial and error to find the startangle i used.
```
